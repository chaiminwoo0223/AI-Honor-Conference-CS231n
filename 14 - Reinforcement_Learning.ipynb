# What is Reinforcement Learning?
## 1.Reinforcement Learning(강화 학습)이란 무엇일까?

- Machine Learning의 한 범주로써 어떤 환경 안에서 정의된 에이전트가 현재의 상태를 인식하여, 
- 선택 가능한 행동들 중 보상을 최대화하는 행동 혹은 행동 순서를 선택하는 방법이다.

image.png

- Agent : Environment에서 Action을 취할 수 있는 물체를 말한다.
- Environment : Agent와 상호작용을 하는 것으로 Agent에게 적절한 state를 부여한다.
-  Agnet는 어떠한 Action에 의해 Reward를 최대로 받는 것이 목적이다.

### 1)순서

- Environmet로 부터 Agent가 State(상태)를 부여받는다.
- 상태를 받은 Agent가 어떠한 Action을 취한다.
- 그 Action에 대해 Agent는 보상을 받는다.
- 그리고 다음 State를 부여받게 된다.

### 2)강화 학습 적용 사례

- 로봇을 앞으로 가도록 하기(Robot Locomotion)
- 게임을 가장 높은 점수로 끝내기(Atari Games)
- 바둑 게임 이기기(Go)

## 1.Markov Property
- 현재 상태로 전체 상태를 나타내는 것을 말한다.
- S : 가능한 상태 집합을 의미
- A : 가능한 액션 집합을 의미
- R : (상태, 액션)의 쌍이 주어졌을 때 보상에 대한 분포
- P : (상태, 액션)의 쌍이 주어졌을 때 다음 상태에 대한 분포(전이 확률)
- R : 보상을 받는 시간에 대해서 얼마나 중요하게 생각하는지를 의미
