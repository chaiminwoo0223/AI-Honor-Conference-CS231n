# VGGNet & GoogleNet

image.png

- 2014년, VGGNet과 GoogleNet이 이미지넷 대회에서 우승
- 기존보다 더 깊은 신경망을 사용
- 2012년과 2013년에는 8계층 -> 2014년에는 19계층, 22계층

## 1.VGGNet

image.png

- VGGNet은 더 깊은 망을 사용하고, 더 작은 필터를 사용
- VGG16(16계층), VGG19(19계층) 
- 매우 작은 필터(3 x 3)와 주기적인 풀링(Pooling)을 망 끝까지 유지

질문 : 왜 더 작은 필터들을 사용할까요? (3 x 3 콘브)
- 3개의 3 x 3 콘브(스트라이드 1) 계층을 쌓으면, 하나의 7 x 7 콘브 계층과 같은 실질적인 수용장을 가짐
- 첫번째 계층에서는 수용장이 3x3이 되며, 두번째 계층에서는 5 x 5가 되며, 세번째 계층에서는 7 x 7이 됨
- 작은 필터를 사용하면, 더 많은 비선형성과 더 적은 파라미터를 가짐

## 2.GoogleNet

image.png

- GoogleNet은 VGG보다 더 깊은 망을 사용(22 계층)
- 계산 효율성을 위해서, 망 아키텍처를 효율적으로 설계 -> 인셉션 모듈(Inception Module)
- GoogleNet은 많은 인셉션 모듈을 서로 쌓아, 많은 파라미터를 줄임
- AexNet이 6000만 개의 파라미터를 가진 것과 비교하면, 전체적으로 500만개의 파라미터를 가짐(1/12)

- 인셉션 모듈 : 좋은 지역 망 위상(local network typology)을 서로 쌓아 올린 것(망 내의 망)
- 인센셥 모듈에서 완전 연결 계층은 없으며, 입력에 대한 다양한 연산을 병렬적으로 수행(합성곱)
- 다양한 필터 연산과 다양한 풀링 연산을 수행

image.png

전체적인 구조
- 스템 망(Setm Network) : 평범한 콘브 망이며, 6개의 시퀀스 계층
- 인센셥 모듈을 쌓음
- 분류기 출력

# Other Architecture

## 1.Network in Network

image.png

- 2014년, 네트워크 안의 네트워크(Network in Network)

- GoogleNet에 대한 철학적 영감을 줌

## 2.FractalNet

image.png

image.png

- 프렉탈(Fractal) : 작은 구조가 전체 구조와 비슷한 형태로 끝없이 되풀이 되는 구조
- 얕은 망에서 깊은 망으로 효율적으로 전이하는 것(Transitioning)
- 출력에 대한 얕은 경로와 깊은 경로가 모두 존재
- 훈련 시, 하위 경로를 드롭아웃(Dropout)
- 테스트 시, 전체 프랙탈 망을 사용

## 3.DenseNet

image.png

- 밀도 높게 연결된 합성곱 망 (Densely Connected Convolutional Network)
- Dense block은 각 계층에서 모든 다른 계층에 피드포워드(FeedForward) 방식으로 연결
- 경사가 사라지는 것을 낮춤 
- 피처 전파를 강화하고, 피처 재사용을 장려

## 4.SqueezeNet

image.png

- 매우 효율적인 망
- Fire 모듈은 1 x 1 필터가 있는 'squeeze' 계층으로 구성되어 있으며, 1 x 1 필터와 3 x 3 필터가 있는 'expand' 계층으로 출력을 전달
- ImageNet보다 50배 적은 파라미터
- AlexNet 수준의 정확도
- AlexNet보다 510배 더 작게 압출할 수 있음(0.5Mb)
