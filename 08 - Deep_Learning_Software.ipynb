# CPU vs GPU

## 1.CPU vs GPU

image.png

1.CPU는 컴퓨터의 뇌에 해당하는 부분이다.
- 컴퓨터에서 구성 단위 중 기억, 해석, 연산, 제어 등을 수행하는 중요한 부분이다.
- 좀 더 적은 Core로 연속적인 일을 처리하는데 자주 사용된다.

2.GPU는 행렬곱 연산에 최적화되어 있으며, 딥러닝 연산에 사용된다.
- 내부적으로 자체적인 RAM을 가지고 있다.
- 더 많은 core로 병렬적인 일을 처리하는데 자주 사용된다.
- 딥러닝 연산을 수행할 때, NVIDIA를 사용한다.

3.GPU Programming
- Cuda(NVIDIA only)
- OpenCL
- Udacity

- HDD 대신 SSD 사용하기
